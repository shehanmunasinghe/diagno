# DxMap

```
def getDxMap(csv_name):
    csv_path = os.path.join(PROJECT_PATH,csv_name)

    rows = open(csv_path).read().strip().split("\n")[1:]
    # random.shuffle(rows)

    print('Total labels: %d'%(len(rows)))

    dx_map ={}
    # for (i, row) in tqdm(enumerate(rows)):
    for row in tqdm(rows):
        description, code, abbr= row.strip().split(",")
        dx_map[code] = (abbr,description)

    return dx_map
```

```
dx_map = getDxMap("Dx_map.csv")
```

# Explore Datasets

```
 from tqdm.notebook import tqdm
```

```
from collections import defaultdict
```

```
# Find unique classes.
def get_classes(input_directory, filenames):
    # classes = set()
    classes = defaultdict(int)
    for filename in tqdm(filenames):
        with open(filename, 'r') as f:
            for l in f:
                if l.startswith('#Dx'):
                    tmp = l.split(': ')[1].split(',')
                    if tmp:
                        for c in tmp:
                            # classes.add(c.strip())
                            c = c.strip()
                            classes[c]+=1
                    else:
                        classes["NoLabel"]+=1
    # return sorted(classes)
    return classes
```

```
def exploreDataset(dset):
    dataset_dir = datasets[dset]

    header_files = []

    lsdir = os.listdir(dataset_dir)
    for f in tqdm(lsdir):
        g = os.path.join(dataset_dir, f)
        if not f.lower().startswith('.') and f.lower().endswith('hea') and os.path.isfile(g):
            header_files.append(g)

    classes = get_classes(dataset_dir, header_files)

    # num_classes = len(classes)
    num_files = len(header_files)
    # recordings = list()
    # headers = list()

    print("Dataset: ",dset)
    # print("num_classes",num_classes)
    print("classes",classes)
    print("num_files",num_files)
    # print("")

    # return classes
    return sorted(classes.items(), key=lambda k_v: k_v[1], reverse=True)
```
